```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

```python
# Load the dataset

df = pd.read_csv("insurance.csv")
```

## Observations

- The dataset contains **1338 records** and **7 columns**.
- There are **4 numerical features**: age, bmi, children, and charges.
- There are **3 categorical features**: sex, smoker, and region.
- The target variable is **charges**, which represents the medical insurance cost.

```python
# Check for missing values in each column

missing_values = df.isnull().sum()

print("Missing Values:")
print(missing_values)
```

```python
# Create a LabelEncoder object

label_encoder = LabelEncoder()
```

```python
# Encode 'sex'
df["sex"] = label_encoder.fit_transform(df["sex"])

# Encode 'smoker'
df["smoker"] = label_encoder.fit_transform(df["smoker"])

# Encode 'region'
df["region"] = label_encoder.fit_transform(df["region"])
```

```python
# Display the first five records after encoding

df.head()
```

```python
# Separate input features (X) and target variable (y)

X = df.drop("charges", axis=1)

y = df["charges"]
```

```python
print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
# Split the dataset into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print("Training Features Shape:", X_train.shape)
print("Testing Features Shape:", X_test.shape)

print("Training Target Shape:", y_train.shape)
print("Testing Target Shape:", y_test.shape)
```

## Observations

- The dataset contains no missing values, so no data cleaning was required.
- The categorical features (`sex`, `smoker`, and `region`) were successfully encoded into numerical values using LabelEncoder.
- The dataset was split into **80% training data (1070 records)** and **20% testing data (268 records)** for model training and evaluation.

# Task 3: Model Development

In this task, we build a Multiple Linear Regression model using the selected features, train it on the training dataset, and predict the insurance charges for the test dataset.

```python
# Create the Multiple Linear Regression model

model = LinearRegression()
```

```python
# Train the model using the training dataset

model.fit(X_train, y_train)
```

```python
# Display model coefficients

coefficients = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": model.coef_
})

coefficients
```

```python
# Display the intercept of the model

print("Intercept:", model.intercept_)
```

```python
# Predict insurance charges for the test dataset

y_pred = model.predict(X_test)
```

```python
# Compare actual and predicted values

comparison = pd.DataFrame({
    "Actual Charges": y_test.values,
    "Predicted Charges": y_pred
})

comparison.head(10)
```

```python
# Predict the insurance charge for the first test sample

sample_prediction = model.predict(X_test.iloc[[0]])

print("Predicted Charge:", sample_prediction[0])
print("Actual Charge:", y_test.iloc[0])
```

## Observations

- The Multiple Linear Regression model was successfully trained using the training dataset.
- The model learned the relationship between customer information and insurance charges.
- Predictions were generated for the test dataset and can now be evaluated using performance metrics.

# Task 4: Model Evaluation

In this task, we evaluate the performance of the Multiple Linear Regression model using Mean Absolute Error (MAE), Mean Squared Error (MSE), and R² Score. We also visualize the relationship between the actual and predicted insurance charges using a scatter plot.

```python
# Calculate Mean Absolute Error

mae = mean_absolute_error(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
```

```python
# Calculate Mean Squared Error

mse = mean_squared_error(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
```

```python
# Calculate R² Score

r2 = r2_score(y_test, y_pred)

print("R² Score:", r2)
```

```python
# Display evaluation metrics

print("Model Evaluation Metrics")
print("-" * 30)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R² Score: {r2:.4f}")
```

```python
# Create a scatter plot of actual vs predicted charges

plt.figure(figsize=(8, 6))

plt.scatter(y_test, y_pred, alpha=0.7)

plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.title("Actual vs Predicted Insurance Charges")

# Add the ideal prediction line
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    color="red",
    linewidth=2
)

plt.grid(True)

plt.show()
```

## Observations

1. The R² Score indicates that the Multiple Linear Regression model explains a significant portion of the variation in medical insurance charges.
2. The scatter plot shows that most predicted values are close to the actual values, indicating a good fit, although some high-cost cases have larger prediction errors.
3. The MAE and MSE values suggest that the model performs reasonably well but still has some prediction error, especially for customers with very high insurance charges.
